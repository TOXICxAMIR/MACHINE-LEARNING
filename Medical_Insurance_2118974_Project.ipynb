```python

def remove_outliers_iqr(data, threshold=1.5):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - threshold * iqr
    upper_bound = q3 + threshold * iqr
    outliers_removed = data[(data >= lower_bound) & (data <= upper_bound)]
    return outliers_removed

insurance_dataset['bmi'] = remove_outliers_iqr(insurance_dataset['bmi'])
```

```python

insurance_dataset['bmi'].plot(kind='box') #USING BOX PLOT TO CHECK IF THERE ANE ANY OUTLIERS PRESENT
```

```python

insurance_dataset.isna().sum()
```

```python

insurance_dataset.dropna(inplace=True)

insurance_dataset.isna().sum()
```

```python

X = insurance_dataset.drop(columns='charges', axis=1) #here in axis 1 represent columns and 0 represent rows
Y = insurance_dataset['charges']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2) 
```

```python

# loading the Linear Regression model
regressor = LinearRegression()

model=regressor.fit(X_train, Y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python

trainscore=regressor.score(X_train, Y_train) #automates the prediction of your data using X_test and compares it with Y_test
trainscore
```

```python

testscore=regressor.score(X_test, Y_test)
testscore
```

```python

difference=trainscore-testscore
difference
```

```python
# R squared value
r2_test = metrics.r2_score(Y_test, y_pred)
print('R squared vale : ', r2_test)
```

```python

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Y_test, y_pred)
mse
```
